```python
import networkx as nx
import matplotlib.pyplot as plt
import sympy as syp

options = {
    "font_size": 36,
    "node_size": 2500,
    "node_color": "lightblue",
    "edgecolors": "black",
    "linewidths": 3,
    "width": 3,
    "font_family": "sans"
}
```

# Graph theory

A graph $G$ is a set of vertices (nodes) $V$ together with a set of edges $E$ that connect the nodes. So if a graph has two nodes $v_1$ and $v_2$ in it's vertex set then we denote that there is an edge between $v_1$ and $v_2$ by having the element $(v_1,v_2)$ in the edge set $E$. For example for the vertex set $V=\{1,2,3,4\}$ and the edge set $E=\{(1,2), (2,3), (1,3), (2,4)\}$ the graph $G := \{V,E\}$ looks like

```python
G = nx.Graph()
V = [1,2,3,4]
E = [(1,2), (2,3), (1,3), (2,4)]
G.add_nodes_from(V)
G.add_edges_from(E)

# explicitly set node positions
pos = {1: (0, 0), 2: (0, -1), 3: (1, 0), 4: (1, -1) }
nx.draw_networkx(G, pos, **options)
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

Graphs can also be *direct* meaning the edges point from one node to another for example the edge $(1,2)$ denotes that there is an edge connecting nodes 1 and 2 that points from node 1 to node 2. The graph $G= {V, E}$ where $V=\{1,2,3,4\}$ and $E=\{(1,2), (2,3), (1,3), (2,4)\}$ is shown below.

```python
DiG = nx.DiGraph()
V = [1,2,3,4]
E = [(1,2), (2,3), (3,1), (4,2)]
DiG.add_nodes_from(V)
DiG.add_edges_from(E)

nx.draw_networkx(DiG, pos, **options)
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

The edges of graphs can also have an attribute associated with it. Most often this attribute is numerical and called the *weight* of an edge.

```python
G_weighted = nx.Graph()
V = [1,2,3,4]
E = [(1, 2, {'weight':2.5}), (2, 3, {'weight':1.2}), (1, 3, {'weight':4.4}), (2, 4, {'weight':3.9})]
G_weighted.add_nodes_from(V)
G_weighted.add_edges_from(E)

# explicitly set node positions
# G[i][j] returns edge data of edge (i,j)
pos = {1: (0, 0), 2: (0, -1), 3: (1, 0), 4: (1, -1) }
nx.draw_networkx(G, pos, **options)
nx.draw_networkx_edge_labels(
    G, pos, 
    edge_labels={ E : G_weighted[E[0]][E[1]]['weight']  for E in G_weighted.edges()}, # loop trough edges and make dict { edge: weight }
    font_color='black',
    font_size=15
)

ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

## Adjacency matrix 

While representing graphs pictorially is useful for visualization it is not straightforward to work with mathematically or in code. Thus it is important to have ways of representing graphs that are easier to work with in this regard. The main way of represent a graph we will go over is by an adjacency matrix. 

For a graph with $n$ nodes the adjacency matrix $A$ of an *undirected* graph $G$ is a $n\times n$ matrix where the element $A_{ij}=1$ if there is an edge between node $i$ and $j$ and $0$ if there is no edge. This immediately implies that the adjacency matrix of undirected graphs is symmetric i.e. $A_{ij}=A_{ji}$. For example the adjacency matrix of the undirected graph shown above is

```python
A = nx.to_numpy_array(G)
syp.Matrix(A)
```

And the adjacency matrix of the directed graph shown above is

```python
A = nx.to_numpy_array(DiG)
syp.Matrix(A)
```

And the adjacency matrix of the weighted undirected graph shown above is

```python
A = nx.to_numpy_array(G_weighted)
syp.Matrix(A)
```

## Graph properties 

### Connectivity 

#### Undirected graphs

For undirected graphs a graph is called connected if there is a path from node $u$ to node $v$ for all pairs $(u,v)\in V$. It's easy to visually see if a graph is not connected since there will be multiple (at least $2$) graphs that share no edge between them in the picture. In that case there are at least $2$ subgraphs which are connected and are called connected components of the graph. 

#### Directed graphs

For directed graphs there are some nuances since it is possible for there to exist a path from node $u$ to node $v$ but *not* one that goes from $v$ to $u$. A directed graph is called *weakly* connected if it's undirected version is connected. The undirected version is created by simply swapping all the edges with undirected ones thereby creating an undirected graph. A directed graph is called *strongly* connected if there is a path from node $u$ to node $v$ *and* a path from node $v$ to node $u$ for all pairs $(u,v)\in V$.

### Trees

A tree is any connected acyclic graph and has the following properties:

1. A tree of $n$ nodes has $n-1$ edges
2. There is exactly 1 path between any two nodes in the tree.
3. Any connected graph with $|V|-1$ edges is a tree.

Note that all three of these statements imply the others, meaning if you can show one to be true for a graph $G$ then the others are also true and $G$ is a tree.
