# Partial trace

## Motivation

The partial trace is a operation that appears in many places in quantum information and computing. It is used when one wants to examine a subsystem of a larger system. In general if one has a quantum system $A$ that is described by vectors in a Hilbert space $\mathcal{H}_A$ and another quantum system $B$ whose state is described by vectors in a Hilbert space $\mathcal{H}_B$ then the joint or composite system of $A$ and $B$, which resides in the joint Hilbert space $\mathcal{H}_A \otimes \mathcal{H}_B$, is given by

$$
\psi_{AB} = \psi_A \otimes \psi_B
$$

and for density matrices the joint density matrix is given similarly by

$$
\rho_{AB} = \rho_A \otimes \rho_B
$$

where $\otimes$ denotes the Kronecker product between the matrices $\rho_A$ and $\rho_B$.

So we have a composite (bipartite) system. Now what if we want to study only one of the subsystems? We need an operation that will take the joint density matrix and return the density matrix of the system we are interested in. 

As a practical example in a recent paper we published {cite:p}`anteneh2024machine` we performed the task of controlling the measurement-based quantum circuit shown below to generate squeezed Schr√∂dinger cat states.

The circuit consists of two qumodes (essentially quantum harmonic oscillators). The first qumode is an arbitrary input state $\rho_j$ and the second qumode is a vacuum state $\ket{0}\bra{0}$ which is transformed into a squeezed vacuum state by the squeezing operation $S(r_j)$. The two qumodes are then acted on by the beam splitter operation $B(\tau_j)$ and afterwards the first qumode is measured in the photon number (Fock) basis. After this we take the non-measured qumode denoted as $\rho_{j+1}$ and feed it back into the circuit to do things all over again. The hope was that we could train an AI to learn what beam splitter and squeezing parameters, $\tau_j$ and $r_j$ respectively, to select such that the output qumode $\rho_{j+1}$ would approximate a squeezed Schr√∂dinger cat state over a finite number of iterations of the circuit. 

The partial trace is used here once we perform the photon number measurement and want to recover the reduced density matrix of the non-measured qumode $\rho_{j+1}$.

Note that the partial trace will always result in a density matrix. This is because the composite system is in general entangled and as we saw in the previous page considering only one subsystem of an entangled state always results in a statistical mixture/mixed state. Also note that the output state $\rho_{j+1}$ of the circuit above is always a pure state, assuming the circuit is ideal/error free, despite the beam splitter operation which in general produces entangled states. This might be confusing at first but this is the case due to the photon number measurement which destroys the entanglement.
